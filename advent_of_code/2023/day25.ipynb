# Part A

```python
data = """jqt: rhn xhk nvd
rsh: frs pzl lsr
xhk: hfx
cmg: qnr nvd lhk bvb
rhn: xhk bvb hfx
bvb: xhk hfx
pzl: lsr hfx nvd
qnr: nvd
ntq: jqt hfx bvb xhk
nvd: lhk
lsr: lhk
rzs: qnr cmg lsr rsh
frs: qnr lhk lsr
"""
```

```python
g = networkx.Graph()
drop = [set(['dsr','xzn']), set(['tbq','qfj']),set(['xbl','qqh'])]

for i, l in enumerate(data.splitlines()):
    a, bs = l.split(': ')
    bs = bs.split()
    for b in bs:
        if set([a,b]) in drop:
            continue
        g.add_edge(a,b)

networkx.draw_spring(g, with_labels=True)
```

```python
total = (len(nx.descendants(g, 'dsr')) + 1) * (len(nx.descendants(g, 'xzn')) + 1)
total
```

```python
submit(total, year=YEAR, day=DAY)
```
