```python
def print_array(x):
    for xx in x:
        print(''.join([str(int(xxx)) for xxx in xx]))
```

```python
explode = np.zeros((2*data.shape[0], 2*data.shape[1]))

for i in range(1, len(path)):
    prev = path[i-1]
    current = path[i]
    
    explode[tuple(current*2)] = 1
    explode[tuple((current + prev))] = 1

# print_array(explode)
```

```python
def mark_neighbors(current):
    for dir in [up, down, left, right]:
        try:
            if explode[tuple(current + dir)] == 0:
                explode[tuple(current + dir)] = 2
                mark_neighbors(current+dir)
        except:
            pass
```

```python
sys.setrecursionlimit(150000)

mark_neighbors(np.array((0,0)))
```
